```python
import json
import pandas as pd
import requests 
```

```python
'''
Here we will try and get as much data as we can on the top 100 venture investors
and the names of the top 100 investors were mentioned on CB insights website and
it was copied into the top_100_VC excel sheet

'''


VCs = pd.read_excel("top_100_VC.xlsx")
VCs_new = []
for i in list(VCs['Name']):
    v = i.split()
    VCs_new.append(v)
```

```python
VC_CB_links = []
one = "https://api.crunchbase.com/v/3/odm-people?name="
three = "%20"
five = "&"
six = "user_key=PUT YOUR CB API USER KEY"


def CB_links():
    for i in VCs_new:
        VC_CB_links.append(one+(i[0])+three+(i[1])+five+six)
CB_links()
```

```python
z =[]
t =[]
CB = []

Crunshbase = {"first_name":[], "last_name":[],"gender":[],"twitter_url":[],
              "facebook_url":[],"linkedin_url":[],"country_code":[],
              "city_name":[],"organization_name":[],"title":[]}
col = ["first_name","last_name","gender","city_name","country_code",
       "title","organization_name","twitter_url","facebook_url","linkedin_url"]

for y in VC_CB_links[0:100]:
    CB = (requests.get(y))  
    CB = json.loads(CB.content)
    CB = CB['data']['items'][0]['properties']
    for i in CB:
        if i == "first_name":
            Crunshbase['first_name'].append(CB[i])
        elif i == "last_name":
            Crunshbase['last_name'].append(CB[i])
        elif i == "gender":
            Crunshbase['gender'].append(CB[i])
        elif i == "city_name":
            Crunshbase['city_name'].append(CB[i])
        elif i == "country_code":
            Crunshbase['country_code'].append(CB[i])
        elif i == "title":
            Crunshbase['title'].append(CB[i])
        elif i == "organization_name":
            Crunshbase['organization_name'].append(CB[i])
        elif i == "twitter_url":
            Crunshbase['twitter_url'].append(CB[i])
        elif i == "facebook_url":
            Crunshbase['facebook_url'].append(CB[i])
        elif i == "linkedin_url":
            Crunshbase['linkedin_url'].append(CB[i])
```

```python
crunshbase = pd.DataFrame(Crunshbase)
col = ["first_name","last_name","gender","city_name","country_code",
       "title","organization_name","twitter_url","facebook_url","linkedin_url"]
crunshbase = crunshbase[col]
crunshbase
```
